```python
import numpy as np
import matplotlib.pyplot as plt
```

# 2D Random walk

We will concider three kinds of 2D random talk.
* $dx$ and $dy$ are uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2. Similarly for $dy$.
* $(dx, dy)=(\cos(\theta), \sin(\theta))$ where $\theta$ is uniformly distributed between $[0, 2\pi]$.

For all kinds of the 2D random walk, you should do the following:
* Plot a single trajectory.
* Plot the distribution of the end point for an ensemble of trajectories.
* Plot the histogram of the $r=x^2_f+y^2_f$ for an ensemble of trajectories.

```python
# dx and dy are uniformly distributed between  [1/2,+1/2]
a=0
b=0
step=1000
positionx=np.zeros(step)
positionx[0]=a
positiony=np.zeros(step)
positiony[0]=b

for i in range (1,step):
    a += np.random.random()-0.5
    b += np.random.random()-0.5
    positionx[i] = a
    positiony[i] = b

plt.figure(figsize=(8, 5))
plt.plot(positionx,positiony)
plt.title("2D Random Walk (1) [-0.5,0.5]")
plt.xlabel('Orders')
plt.ylabel('Position')


runs = 1000
final = np.zeros(runs)
positionxx=np.zeros(runs)
positionyy=np.zeros(runs)

for j in range (1,runs):
    positionx = 0
    positiony = 0
    for i in range (1,step):
        positionx += np.random.random(step)-0.5
        positiony += np.random.random(step)-0.5
    positionxx[j] = positionx[i]
    positionyy[j] = positiony[i]
    positionxy = np.sqrt(np.square(positionx)+np.square(positiony))
    final[j] = positionxy[i]

plt.figure(figsize=(8, 5))
plt.plot(positionxx,positionyy,'.')
plt.title("2D Random Walk distribution (1) [-0.5,0.5]")
plt.xlabel('Orders')
plt.ylabel('Position')

plt.figure(figsize=(8, 5))
plt.hist(final,bins=50)
plt.title("2D Random Walk after 1000 runs (1) [-0.5,0.5]")
plt.xlabel('Position')
plt.ylabel('Times')
```

```python
# =+1  with probability 1/2 and =1 with probability 1/2. Similarly for 
c=0
d=0
step=1000
positionx=np.zeros(step)
positionx[0]=c
positiony=np.zeros(step)
positiony[0]=d

for i in range (1,step):
    c += np.random.randint(0,2)*2-1
    d += np.random.randint(0,2)*2-1
    positionx[i] = c
    positiony[i] = d

plt.figure(figsize=(8, 5))
plt.plot(positionx,positiony)
plt.title("2D Random Walk (2) dx=1 or dx=-1/dy=1 or dy=-1")
plt.xlabel('Orders')
plt.ylabel('Position')


runs = 1000
final = np.zeros(runs)
positionxx=np.zeros(runs)
positionyy=np.zeros(runs)

for j in range (1,runs):
    positionx = 0
    positiony = 0
    for i in range (1,step):
        positionx += np.random.randint(0,2,step)*2-1
        positiony += np.random.randint(0,2,step)*2-1
    positionxx[j] = positionx[i]
    positionyy[j] = positiony[i]
    positionxy = np.sqrt(np.square(positionx)+np.square(positiony))
    final[j] = positionxy[i]

plt.figure(figsize=(8, 5))
plt.plot(positionxx,positionyy,'.')
plt.title("2D Random Walk distribution (2) dx=1 or dx=-1/dy=1 or dy=-1")
plt.xlabel('Orders')
plt.ylabel('Position')

plt.figure(figsize=(8, 5))
plt.hist(final,bins=50)
plt.title("2D Random Walk after 1000 runs (2) dx=1 or dx=-1/dy=1 or dy=-1")
plt.xlabel('Position')
plt.ylabel('Times')
```

```python
# (,)=(cos(),sin())  where    is uniformly distributed between  [0,2]
e=0
f=0
step=1000
positionx=np.zeros(step)
positionx[0]=e
positiony=np.zeros(step)
positiony[0]=f


for i in range (1,step):
    theta=np.random.random()*2*np.pi
    e += np.cos(theta)
    f += np.sin(theta)
    positionx[i] = e
    positiony[i] = f

plt.figure(figsize=(8, 5))
plt.plot(positionx,positiony)
plt.title("2D Random Walk (3) (dx,dy)=(cos,sin)")
plt.xlabel('Orders')
plt.ylabel('Position')


runs = 1000
final = np.zeros(runs)
positionxx=np.zeros(runs)
positionyy=np.zeros(runs)

for j in range (1,runs):
    positionx = 0
    positiony = 0
    for i in range (1,step):
        theta=np.random.random(step)*2*np.pi
        positionx += np.cos(theta)
        positiony += np.sin(theta)
    positionxx[j] = positionx[i]
    positionyy[j] = positiony[i]
    positionxy = np.sqrt(np.square(positionx)+np.square(positiony))
    final[j] = positionxy[i]

plt.figure(figsize=(8, 5))
plt.plot(positionxx,positionyy,'.')
plt.title("2D Random Walk distribution (3) (dx,dy)=(cos,sin)")
plt.xlabel('Orders')
plt.ylabel('Position')

plt.figure(figsize=(8, 5))
plt.hist(final,bins=50)
plt.title("2D Random Walk after 1000 runs (3) (dx,dy)=(cos,sin)")
plt.xlabel('Position')
plt.ylabel('Times')
```
