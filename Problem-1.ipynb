```python
import numpy as np
import matplotlib.pyplot as plt
```

# 1D Random walk

We will concider two kinds of 1D random talk.
* $dx$ is uniformly distributed between $[-1/2, +1/2]$.
* $dx=+1$ with probability 1/2 and $dx=-1$ with probability 1/2.

For both kinds of the 1D random walk, you should do the following:
* Plot a single trajectory.
* Plot the histogram of the end point for an ensemble of trajectories.

```python
#第一小題 dx is uniformly distributed between  [−1/2,+1/2]
a=0
step=1000
position=np.zeros(step)
position[0]=a

for i in range (1,step):
    a += np.random.random()-0.5
    position[i] = a

plt.figure(figsize=(8, 5))
plt.plot(position)
plt.title("1D Random Walk (1) [-0.5,0.5]")
plt.xlabel('Orders')
plt.ylabel('Position')

runs = 1000
final = np.zeros(runs)

for j in range (1,runs):
    position =0
    for i in range (1,step):
        position += np.random.random(step)-0.5
    final[j] = position[i]
        
plt.figure(figsize=(8, 5))
plt.hist(final,bins=40)
plt.title("1D Random Walk after 1000 runs (1) (-0.5,0.5)")
plt.xlabel('Position')
plt.ylabel('Times')
```

```python
#第二小題 𝑑𝑥=+1  with probability 1/2 and 𝑑𝑥=−1 with probability 1/2.
b=0
step=1000
position2=np.zeros(step)
position2[0]=b

for i in range (1,step):
    b += np.random.randint(0,2)*2-1
    position2[i] = b

plt.figure(figsize=(8, 5))
plt.plot(position2)
plt.title("1D Random Walk (2) dx=1 or dx=-1")
plt.xlabel('Orders')
plt.ylabel('Position')

runs = 1000
final = np.zeros(runs)

for j in range (1,runs):
    position2 =0
    for i in range (1,step):
        position2 += np.random.randint(0,2,step)*2-1
    final[j] = position2[i]
        
plt.figure(figsize=(8, 5))
plt.hist(final,bins=40)
plt.title("1D Random Walk after 1000 runs (2) dx=1 or dx=-1")
plt.xlabel('Position')
plt.ylabel('Times')
```
